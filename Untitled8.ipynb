Step 1: Bootstrap Sampling and Slope Coefficients
We create bootstrapped samples of the "long wait times" data, fit a simple linear regression for each sample, and collect the slope coefficients.

Step 2: Simulate Data Under the Null Hypothesis
We simulate the data assuming no linear association (slope = 0), fit simple linear regressions for these simulated datasets, and collect the slope coefficients.

Step 3: Calculate Confidence Interval and P-value
We:

Calculate the 95% bootstrapped confidence interval for the slope coefficients.
Compare the observed slope coefficient to the simulated null hypothesis distribution and compute the p-value.

Step 4: Visualization
We visualize the distributions of the slope coefficients for both bootstrapped and simulated samples.

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
from scipy import stats

# Filter long wait times
long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit
long_wait_data = old_faithful[long_wait]

# Observed model
observed_model = smf.ols('duration ~ waiting', data=long_wait_data).fit()
observed_slope = observed_model.params[1]

# Step 1: Bootstrap Sampling
bootstrapped_slope_coefficients = []

for _ in range(1000):  # 1000 bootstrap samples
    sample_data = long_wait_data.sample(n=len(long_wait_data), replace=True)
    bootstrap_model = smf.ols('duration ~ waiting', data=sample_data).fit()
    bootstrapped_slope_coefficients.append(bootstrap_model.params[1])

# Step 2: Simulated Data Under Null Hypothesis
simulated_slope_coefficients = []

# Create the simulated dataset
old_faithful_simulation = long_wait_data.copy()
for _ in range(1000):  # 1000 simulations
    old_faithful_simulation['duration'] = (
        1.65 + 0 * old_faithful_simulation.waiting +
        stats.norm(loc=0, scale=0.37).rvs(size=len(old_faithful_simulation))
    )
    simulated_model = smf.ols('duration ~ waiting', data=old_faithful_simulation).fit()
    simulated_slope_coefficients.append(simulated_model.params[1])

# Step 3: Confidence Interval and P-value
# Bootstrapped Confidence Interval
bootstrapped_confidence_interval = np.quantile(bootstrapped_slope_coefficients, [0.025, 0.975])

# Simulated P-value
simulated_p_value = (np.abs(simulated_slope_coefficients) >= np.abs(observed_slope)).mean()

# Step 4: Visualization
# Bootstrapped Slopes
fig_bootstrap = px.histogram(bootstrapped_slope_coefficients, nbins=50,
                             title="Bootstrapped Sampling Distribution of Slope Coefficients",
                             labels={'value': 'Slope Coefficient', 'count': 'Frequency'})
fig_bootstrap.add_vline(x=observed_slope, line_dash="dash", line_color="red", annotation_text="Observed Slope")

# Simulated Slopes
fig_simulation = px.histogram(simulated_slope_coefficients, nbins=50,
                              title="Simulated Sampling Distribution of Slope Coefficients (Null Hypothesis)",
                              labels={'value': 'Slope Coefficient', 'count': 'Frequency'})
fig_simulation.add_vline(x=observed_slope, line_dash="dash", line_color="red", annotation_text="Observed Slope")

# Show results
print("Observed Slope Coefficient:", observed_slope)
print("Bootstrapped 95% Confidence Interval:", bootstrapped_confidence_interval)
print("Simulated P-value:", simulated_p_value)

fig_bootstrap.show()
fig_simulation.show()
```

The "short" and "long" wait times are not "before and after" measurements so there are not natural pairs on which to base differences on which to do a "one sample" (paired differences) hypothesis test; but, we can do "two sample" hypothesis testing using a permuation test, or create a 95% bootstrap confidence interval for the difference in means of the two populations.
Task (a): Explain Sampling Approaches
Permutation Test Sampling:

In the permutation test, labels (kind) are shuffled randomly, assuming the null hypothesis that the means of the two groups are equal.
By recalculating the mean difference statistic (mean of short minus long) for each shuffle, we generate a distribution of the test statistic under the null hypothesis.
The observed mean difference is compared to this null distribution to compute a p-value.
Bootstrap Sampling:

Bootstrap sampling resamples the data within each group (short and long) independently, with replacement, to mimic the sampling variability.
For each bootstrap sample, the mean difference statistic is calculated.
The 95% confidence interval is constructed by taking the 2.5th and 97.5th percentiles of the bootstrapped distribution of mean differences.

Task (b): Compare with Indicator Variable Model
The permutation test and bootstrap confidence interval methods are non-parametric, making fewer assumptions about the data distribution compared to the indicator variable model, which assumes a linear relationship between predictors and outcomes.
Similarity:
Both methods evaluate differences between groups.
Difference:
Indicator model estimates the difference in means using a regression framework with assumptions about normality and homoscedasticity.
Permutation and bootstrap methods directly calculate differences empirically without assuming any underlying distribution.
